```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/healthcare-dataset-stroke-data.csv')
```

```python
data
```

```python
#Question 2
#variable 1: Age

mean = data['age'].mean()
print("mean for age:", (round(mean,2)))

median = data['age'].median()
print("median for age:", (round(median,2)))

mode= data['age'].mode()
print("mode for age:", (round(mode,2)))

std_dev = data['age'].std()
print("standard deviation for age:", (round(std_dev,2)))
```

```python
#Question 2
#Variable 2: Average glucose level

mean = data['avg_glucose_level'].mean()
print("mean for average glucose level:", (round(mean,2)))

median = data['avg_glucose_level'].median()
print("median for average glucose level:", (round(median,2)))

mode = data['avg_glucose_level'].mode()
print("mode for average glucose level:", (round(mode,2)))

std_dev = data['avg_glucose_level'].std()
print("standard deviation for average glucose level:", (round(std_dev,2)))
```

```python
#Question 2
#Variable 3: bmi

mean = data['bmi'].mean()
print("mean for bmi:", (round(mean,2)))

median = data['bmi'].median()
print("median for bmi:", (round(median,2)))

mode = data['bmi'].mode()
print("mode for bmi:", (round(mode,2)))

std_dev = data['bmi'].std()
print("standard deviation for bmi:", (round(std_dev,2)))
```
